````
import os, shutil, random, glob
from pathlib import Path

RAW = "/content/ds/extracted/PSORIASIS AND NORMAL SKIN"  # detected from your output

DATA_ROOT = "/content/psoriasis_data"
TRAIN_DIR = f"{DATA_ROOT}/train"
VAL_DIR   = f"{DATA_ROOT}/val"
TEST_DIR  = f"{DATA_ROOT}/test"

EXTS = ["*.jpg","*.jpeg","*.png","*.bmp","*.tif","*.tiff","*.webp","*.pgm","*.ppm"]

def list_images(root):
    imgs = []
    for ext in EXTS:
        imgs += glob.glob(str(Path(root)/"**"/ext), recursive=True)
    return imgs

def safe_reset_data_root():
    if os.path.exists(DATA_ROOT):
        shutil.rmtree(DATA_ROOT)
    for split in [TRAIN_DIR, VAL_DIR, TEST_DIR]:
        for cls in ["psoriasis","normal"]:
            os.makedirs(f"{split}/{cls}", exist_ok=True)

def print_counts():
    def cnt(d): return len(list_images(d))
    print("Counts per split/class:")
    for split, path in [("train",TRAIN_DIR), ("val",VAL_DIR), ("test",TEST_DIR)]:
        print(f"- {split}: psoriasis={cnt(os.path.join(path,'psoriasis'))}, normal={cnt(os.path.join(path,'normal'))}")

def find_class_dirs(root):
    # Attempt to find class folders inside RAW
    subs = [p for p in Path(root).glob("*") if p.is_dir()]
    # map names to lower for matching
    mapping = {"psoriasis": None, "normal": None}
    for sd in subs:
        name = sd.name.lower()
        if "psor" in name or "psoriasis" in name or "plaque" in name or "lesion" in name:
            mapping["psoriasis"] = str(sd)
        if "norm" in name or "healthy" in name or "control" in name:
            mapping["normal"] = str(sd)
    return mapping

def copy_split(files, split_dir, start, end, cls_name):
    for p in files[start:end]:
        shutil.copy2(p, f"{split_dir}/{cls_name}/{os.path.basename(p)}")

# Build splits
safe_reset_data_root()
mapping = find_class_dirs(RAW)

if mapping["psoriasis"] and mapping["normal"]:
    ps_files = list_images(mapping["psoriasis"])
    nm_files = list_images(mapping["normal"])
    random.shuffle(ps_files); random.shuffle(nm_files)

    def split_and_copy(files, cls):
        n = len(files)
        tr, va = int(0.8*n), int(0.9*n)
        copy_split(files, TRAIN_DIR, 0, tr, cls)
        copy_split(files, VAL_DIR, tr, va, cls)
        copy_split(files, TEST_DIR, va, n, cls)

    split_and_copy(ps_files, "psoriasis")
    split_and_copy(nm_files, "normal")
else:
    # Fallback: keyword routing by filename/path
    psoriasis_kw = {"psoriasis","psoriatic","plaque","lesion"}
    normal_kw    = {"normal","healthy","control"}
    def infer_cls(p):
        low = p.lower()
        if any(k in low for k in psoriasis_kw): return "psoriasis"
        if any(k in low for k in normal_kw):    return "normal"
        return "normal"
    all_imgs = list_images(RAW)
    random.shuffle(all_imgs)
    by_cls = {"psoriasis":[], "normal":[]}
    for p in all_imgs:
        by_cls[infer_cls(p)].append(p)
    for cls, files in by_cls.items():
        n = len(files)
        tr, va = int(0.8*n), int(0.9*n)
        copy_split(files, TRAIN_DIR, 0, tr, cls)
        copy_split(files, VAL_DIR, tr, va, cls)
        copy_split(files, TEST_DIR, va, n, cls)

print_counts()
print("TRAIN_DIR =", TRAIN_DIR)
print("VAL_DIR   =", VAL_DIR)
print("TEST_DIR  =", TEST_DIR)
````

````
# ===== Config =====
import torch, os

class Cfg:
    TRAIN_DIR = "/content/psoriasis_data/train"
    VAL_DIR   = "/content/psoriasis_data/val"
    TEST_DIR  = "/content/psoriasis_data/test"
    OUT_DIR   = "/content/outputs"
    os.makedirs(OUT_DIR, exist_ok=True)

    # Start with sanity settings; well switch to full later
    IMAGE_SIZE = (224, 224)
    NUM_CLASSES = 2
    NUM_WORKERS = 2
    PIN_MEMORY  = True

    NORMALIZE_MEAN = [0.485, 0.456, 0.406]
    NORMALIZE_STD  = [0.229, 0.224, 0.225]

    MIXED_PRECISION = True
    PRETRAINED = True
    DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

    PHASES = {
        "phase1": {"epochs": 2, "lr": 1e-3, "batch": 32},  # sanity
        "phase2": {"epochs": 2, "lr": 5e-4, "batch": 32},  # sanity
    }

cfg = Cfg()
print("Device:", cfg.DEVICE, " Image:", cfg.IMAGE_SIZE)

# ===== Albumentations + ImageFolder wrappers =====
import albumentations as A
from albumentations.pytorch import ToTensorV2
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader

def get_transforms(cfg, strong=True):
    train_tf = [
        A.Resize(cfg.IMAGE_SIZE[0], cfg.IMAGE_SIZE[1]),
        A.HorizontalFlip(p=0.5),
        A.RandomBrightnessContrast(0.2, 0.2, p=0.5),
        A.Normalize(mean=cfg.NORMALIZE_MEAN, std=cfg.NORMALIZE_STD),
        ToTensorV2()
    ]
    # For sanity, keep light aug; well enable stronger after this
    val_tf = A.Compose([
        A.Resize(cfg.IMAGE_SIZE[0], cfg.IMAGE_SIZE[1]),
        A.Normalize(mean=cfg.NORMALIZE_MEAN, std=cfg.NORMALIZE_STD),
        ToTensorV2()
    ])
    return A.Compose(train_tf), val_tf

class AlbuWrapper:
    def __init__(self, tfm): self.tfm = tfm
    def __call__(self, img):
        import numpy as np
        img = img.convert("RGB")
        arr = np.array(img)
        return self.tfm(image=arr)["image"]

def make_loaders(cfg, strong_aug=False, phase="phase1"):
    t_tfm, v_tfm = get_transforms(cfg, strong=strong_aug)
    train_ds = ImageFolder(cfg.TRAIN_DIR, transform=AlbuWrapper(t_tfm))
    val_ds   = ImageFolder(cfg.VAL_DIR,   transform=AlbuWrapper(v_tfm))
    bs = cfg.PHASES[phase]["batch"]
    train_loader = DataLoader(train_ds, batch_size=bs, shuffle=True,  num_workers=cfg.NUM_WORKERS, pin_memory=cfg.PIN_MEMORY, drop_last=True)
    val_loader   = DataLoader(val_ds,   batch_size=bs, shuffle=False, num_workers=cfg.NUM_WORKERS, pin_memory=cfg.PIN_MEMORY)
    return train_loader, val_loader, len(train_ds.classes)

# Quick check
train_loader, val_loader, ncls = make_loaders(cfg, strong_aug=False, phase="phase1")
print("Classes detected:", ncls)
print("Train batches:", len(train_loader), " Val batches:", len(val_loader))
````

````
# ===== Models (timm) =====
import timm
import torch.nn as nn

def build_densenet201(cfg):
    m = timm.create_model("densenet201", pretrained=cfg.PRETRAINED, num_classes=cfg.NUM_CLASSES)
    return m

# ===== Eval and training loops =====
from torch.cuda.amp import autocast, GradScaler
import torch.optim as optim
import numpy as np
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score
import torch

def evaluate(model, loader, device):
    model.eval()
    y_true, y_prob = [], []
    with torch.no_grad():
        for x, y in loader:
            x = x.to(device)
            logits = model(x)
            prob = torch.softmax(logits, dim=1)[:, 1].detach().cpu().numpy()
            y_prob.extend(prob.tolist())
            y_true.extend(y.numpy().tolist())
    y_pred = (np.array(y_prob) >= 0.5).astype(int)
    acc = accuracy_score(y_true, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
    try:
        auc = roc_auc_score(y_true, y_prob)
    except:
        auc = float("nan")
    return {"acc":acc, "prec":prec, "rec":rec, "f1":f1, "auc":auc}

def train_phase(model, train_loader, val_loader, device, epochs, lr, use_amp=True):
    model = model.to(device)
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    scaler = GradScaler(enabled=(use_amp and device.type=="cuda"))
    best = {"acc":0.0}
    for ep in range(1, epochs+1):
        model.train()
        running = 0.0
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad(set_to_none=True)
            if scaler.is_enabled():
                with autocast():
                    logits = model(xb)
                    loss = nn.CrossEntropyLoss()(logits, yb)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()
            else:
                logits = model(xb)
                loss = nn.CrossEntropyLoss()(logits, yb)
                loss.backward()
                optimizer.step()
            running += loss.item() * xb.size(0)
        train_loss = running / len(train_loader.dataset)
        metrics = evaluate(model, val_loader, device)
        if metrics["acc"] > best.get("acc", 0):
            best = {"ep":ep, **metrics}
            torch.save(model.state_dict(), "/content/outputs/best.pth")
        print(f"Epoch {ep}/{epochs} | loss {train_loss:.4f} | val acc {metrics['acc']:.4f} "
              f"prec {metrics['prec']:.4f} rec {metrics['rec']:.4f} auc {metrics['auc']:.4f}")
    return best
````

````
# Use the loaders we created earlier (phase1, light aug)
train_loader, val_loader, _ = make_loaders(cfg, strong_aug=False, phase="phase1")

model = build_densenet201(cfg)
best_sanity = train_phase(
    model,
    train_loader, val_loader,
    torch.device(cfg.DEVICE),
    epochs=cfg.PHASES["phase1"]["epochs"],  # 2
    lr=cfg.PHASES["phase1"]["lr"],          # 1e-3
    use_amp=cfg.MIXED_PRECISION
)
print("Sanity best:", best_sanity)
````

````
# Compatible AMP training loop (works on your torch)
from torch.cuda.amp import autocast as cuda_autocast, GradScaler as CudaGradScaler
import torch, torch.nn as nn, torch.optim as optim, numpy as np
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score
from contextlib import nullcontext

def evaluate_eval(model, loader, device):
    model.eval()
    y_true, y_prob = [], []
    with torch.no_grad():
        for x, y in loader:
            x = x.to(device)
            logits = model(x)
            prob = torch.softmax(logits, dim=1)[:, 1].detach().cpu().numpy()
            y_prob.extend(prob.tolist()); y_true.extend(y.numpy().tolist())
    y_pred = (np.array(y_prob) >= 0.5).astype(int)
    acc = accuracy_score(y_true, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
    try: auc = roc_auc_score(y_true, y_prob)
    except: auc = float("nan")
    return {"acc":acc, "prec":prec, "rec":rec, "f1":f1, "auc":auc}

def train_phase_amp(model, train_loader, val_loader, device, epochs, lr, use_amp=True):
    model = model.to(device)
    optimizer = optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    scaler = CudaGradScaler(enabled=(use_amp and device.type=="cuda"))
    autocast_ctx = (lambda: cuda_autocast()) if (use_amp and device.type=="cuda") else nullcontext

    best = {"acc":0.0}
    for ep in range(1, epochs+1):
        model.train()
        running = 0.0
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad(set_to_none=True)
            if scaler.is_enabled():
                with autocast_ctx():
                    logits = model(xb)
                    loss = nn.CrossEntropyLoss()(logits, yb)
                scaler.scale(loss).backward()
                scaler.step(optimizer)
                scaler.update()
            else:
                logits = model(xb)
                loss = nn.CrossEntropyLoss()(logits, yb)
                loss.backward()
                optimizer.step()
            running += loss.item() * xb.size(0)

        train_loss = running / len(train_loader.dataset)
        metrics = evaluate_eval(model, val_loader, device)
        if metrics["acc"] > best.get("acc", 0):
            best = {"ep":ep, **metrics}
            torch.save(model.state_dict(), "/content/outputs/best.pth")
        print(f"Epoch {ep}/{epochs} | loss {train_loss:.4f} | val acc {metrics['acc']:.4f} "
              f"prec {metrics['prec']:.4f} rec {metrics['rec']:.4f} auc {metrics['auc']:.4f}")
    return best
````

````
# Build and train EfficientNet-B7 (timm)
import timm, torch, torch.nn as nn, torch.optim as optim
from torch.cuda.amp import autocast, GradScaler
from contextlib import nullcontext
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score
import numpy as np

def build_efficientnet_b7(cfg):
    return timm.create_model("tf_efficientnet_b7_ns", pretrained=cfg.PRETRAINED, num_classes=cfg.NUM_CLASSES)

def evaluate_eval(model, loader, device):
    model.eval()
    y_true, y_prob = [], []
    with torch.no_grad():
        for x, y in loader:
            x = x.to(device)
            logits = model(x)
            prob = torch.softmax(logits, dim=1)[:, 1].detach().cpu().numpy()
            y_prob.extend(prob.tolist()); y_true.extend(y.numpy().tolist())
    y_pred = (np.array(y_prob) >= 0.5).astype(int)
    acc = accuracy_score(y_true, y_pred)
    prec, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
    try: auc = roc_auc_score(y_true, y_prob)
    except: auc = float("nan")
    return {"acc":acc, "prec":prec, "rec":rec, "f1":f1, "auc":auc}

def train_model_amp(model, train_loader, val_loader, device, epochs, lr, use_amp=True, ckpt="/content/outputs/effb7_best.pth"):
    model = model.to(device)
    opt = optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    scaler = GradScaler(enabled=(use_amp and device.type=="cuda"))
    ac = (lambda: autocast()) if (use_amp and device.type=="cuda") else nullcontext

    best = {"acc":0.0}
    for ep in range(1, epochs+1):
        model.train()
        run = 0.0
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            opt.zero_grad(set_to_none=True)
            if scaler.is_enabled():
                with ac():
                    logits = model(xb)
                    loss = nn.CrossEntropyLoss()(logits, yb)
                scaler.scale(loss).backward(); scaler.step(opt); scaler.update()
            else:
                logits = model(xb); loss = nn.CrossEntropyLoss()(logits, yb)
                loss.backward(); opt.step()
            run += loss.item() * xb.size(0)
        tr_loss = run / len(train_loader.dataset)
        metrics = evaluate_eval(model, val_loader, device)
        if metrics["acc"] > best.get("acc", 0):
            best = {"ep":ep, **metrics}; torch.save(model.state_dict(), ckpt)
        print(f"[EffB7] Epoch {ep}/{epochs} | loss {tr_loss:.4f} | val acc {metrics['acc']:.4f} "
              f"prec {metrics['prec']:.4f} rec {metrics['rec']:.4f} auc {metrics['auc']:.4f}")
    return best

# Use same full settings as DenseNet final (384px, strong aug loaders)
# Reuse make_loaders_full(cfg, phase) from before
train_loader_e, val_loader_e = make_loaders_full(cfg, phase="phase2")
model_e = build_efficientnet_b7(cfg)
best_eff = train_model_amp(model_e, train_loader_e, val_loader_e, torch.device(cfg.DEVICE),
                           epochs=8, lr=5e-4, use_amp=cfg.MIXED_PRECISION, ckpt="/content/outputs/effb7_best.pth")
print("EffB7 best:", best_eff)
torch.save(model_e.state_dict(), "/content/outputs/effb7_final.pth")
````

````
import numpy as np, torch

def prob_model(model, loader, device):
    model.eval()
    ps = []
    with torch.no_grad():
        for x, _ in loader:
            x = x.to(device)
            logits = model(x)
            ps.append(torch.softmax(logits, dim=1)[:,1].cpu())
    return torch.cat(ps)

# Rebuild eval loader (val) with eval tfm
# Reuse make_loaders_full(cfg, "phase2") for val loader batch size
_, val_loader_eval = make_loaders_full(cfg, phase="phase2")

device = torch.device(cfg.DEVICE)
# Load models
m_dn = build_densenet201(cfg); m_dn.load_state_dict(torch.load("/content/outputs/densenet201_final.pth", map_location="cpu")); m_dn=m_dn.to(device)
m_e  = build_efficientnet_b7(cfg); m_e.load_state_dict(torch.load("/content/outputs/effb7_final.pth", map_location="cpu")); m_e=m_e.to(device)

# Collect probabilities
p_dn = prob_model(m_dn, val_loader_eval, device).numpy()
p_e  = prob_model(m_e,  val_loader_eval, device).numpy()

# Simple equal-weight ensemble (you can weight by val acc if desired)
ens_prob = (p_dn + p_e) / 2.0

# Build y_true
y_true = []
for _, y in val_loader_eval:
    y_true.extend(y.numpy().tolist())
y_true = np.array(y_true)
y_pred = (ens_prob >= 0.5).astype(int)

from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score
acc = accuracy_score(y_true, y_pred)
prec, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
auc = roc_auc_score(y_true, ens_prob)
print(f"Ensemble VAL -> acc {acc:.4f} prec {prec:.4f} rec {rec:.4f} f1 {f1:.4f} auc {auc:.4f}")
````

````
# Apply simple TTA (hflip) and average predictions
def prob_model_tta(model, loader, device):
    model.eval()
    probs = []
    with torch.no_grad():
        for x, _ in loader:
            x = x.to(device)
            # original
            logits1 = model(x)
            p1 = torch.softmax(logits1, dim=1)[:,1]
            # hflip
            x2 = torch.flip(x, dims=[3])
            logits2 = model(x2)
            p2 = torch.softmax(logits2, dim=1)[:,1]
            probs.append(((p1+p2)/2.0).cpu())
    return torch.cat(probs)

ens_prob_tta = (prob_model_tta(m_dn, val_loader_eval, device).numpy() + prob_model_tta(m_e, val_loader_eval, device).numpy())/2.0
y_pred_tta = (ens_prob_tta >= 0.5).astype(int)
acc = accuracy_score(y_true, y_pred_tta)
prec, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred_tta, average="binary", zero_division=0)
auc = roc_auc_score(y_true, ens_prob_tta)
print(f"Ensemble VAL (TTA) -> acc {acc:.4f} prec {prec:.4f} rec {rec:.4f} f1 {f1:.4f} auc {auc:.4f}")
````

````
# Fixed Grad-CAM utility: keep input strictly float32
import torch
import torch.nn as nn
import cv2, numpy as np, matplotlib.pyplot as plt
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.image import show_cam_on_image

def get_last_conv_layer(module: nn.Module):
    last_conv = None
    for m in module.modules():
        if isinstance(m, nn.Conv2d):
            last_conv = m
    if last_conv is None:
        raise RuntimeError("No Conv2d layer found in the model for Grad-CAM.")
    return last_conv

def gradcam_on_sample(model, img_path, cfg):
    model.eval()

    im = cv2.imread(img_path)[:, :, ::-1]              # RGB
    im_res = cv2.resize(im, cfg.IMAGE_SIZE[::-1])
    im_norm = (im_res / 255.0).astype(np.float32)      # float32

    mean = np.array(cfg.NORMALIZE_MEAN, dtype=np.float32)
    std  = np.array(cfg.NORMALIZE_STD,  dtype=np.float32)
    im_input = (im_norm - mean) / std                   # float32

    im_t = torch.from_numpy(im_input).permute(2, 0, 1).unsqueeze(0)
    im_t = im_t.to(cfg.DEVICE, dtype=torch.float32)     # ensure float32 on device

    target_layers = [get_last_conv_layer(model)]

    # Use context manager to clean hooks (prevents __del__ warnings)
    with GradCAM(model=model, target_layers=target_layers) as cam:
        grayscale_cam = cam(input_tensor=im_t)[0]

    vis = show_cam_on_image(im_norm, grayscale_cam, use_rgb=True)
    plt.figure(figsize=(5,5)); plt.imshow(vis); plt.axis("off"); plt.show()
````
